```python
def weighted_crosstab(df, row_var, col_var, weight_var='sample_weight'):
    """
    Creates a weighted crosstab as percentages.
    Accounts for survey design weights.
    """
    # Drop missing
    temp = df[[row_var, col_var, weight_var]].dropna()
    
    # Create weighted counts
    weighted = temp.groupby([row_var, col_var])[weight_var].sum().unstack()
    
    # Convert to row percentages
    weighted_pct = weighted.div(weighted.sum(axis=1), axis=0) * 100
    
    return weighted_pct
 
print("=" * 60)
print("WEIGHTED CROSSTAB: Warning Recall vs Daily Use")
print("=" * 60)
print(weighted_crosstab(saw_warnings, 'warn_recall_thcmh', 'daily_dv'))
 
print("\n" + "=" * 60)
print("WEIGHTED CROSSTAB: Warning Recall vs Workplace Use")
print("=" * 60)
print(weighted_crosstab(saw_warnings, 'warn_recall_thcmh', 'work_use_binary'))
 
print("\n" + "=" * 60)
print("WEIGHTED CROSSTAB: Warning Recall vs Mental Health Impact")
print("=" * 60)
# Binary harmful vs not
saw_warnings['harmful_mh'] = saw_warnings['impact_mental'].isin([4.0, 5.0]).astype(float)
print(weighted_crosstab(saw_warnings, 'warn_recall_thcmh', 'harmful_mh'))
```

```python
print("\n" + "=" * 60)
print("WEIGHTED LOGISTIC REGRESSION: Predicting Daily Use")
print("=" * 60)
 
# Prepare clean dataset for regression
reg_vars = ['daily_dv', 'warn_recall_thcmh', 'age6', 'sex', 
            'canpurpose_dv', 'use_modes_smoke', 'sample_weight']
 
reg_data = saw_warnings[reg_vars].dropna().copy()
 
# Fit weighted logistic regression using GLM
X = reg_data[['warn_recall_thcmh', 'age6', 'sex', 'canpurpose_dv', 'use_modes_smoke']]
X = sm.add_constant(X)
y = reg_data['daily_dv']
weights = reg_data['sample_weight']

reg_data['normalized_weight'] = (
    reg_data['sample_weight'] / reg_data['sample_weight'].mean()
)

model = sm.GLM(
    y, X,
    family=sm.families.Binomial(),
    var_weights=reg_data['normalized_weight']
).fit()

print(model.summary())
 

 
# Extract odds ratios
print("\n" + "=" * 60)
print("ODDS RATIOS (easier to interpret)")
print("=" * 60)
odds_ratios = pd.DataFrame({
    'OR': np.exp(model.params),
    'Lower 95% CI': np.exp(model.conf_int()[0]),
    'Upper 95% CI': np.exp(model.conf_int()[1]),
    'p-value': model.pvalues
})
print(odds_ratios.round(3))
```

```python
print("\n" + "=" * 60)
print("WEIGHTED LOGISTIC REGRESSION: Predicting Workplace Use")
print("=" * 60)
 
reg_vars2 = ['work_use_binary', 'warn_recall_thcmh', 'age6', 'sex',
             'canpurpose_dv', 'use_modes_smoke', 'sample_weight']
 
reg_data2 = saw_warnings[reg_vars2].dropna().copy()
 
# Fit weighted logistic regression for workplace use
X2 = reg_data2[['warn_recall_thcmh', 'age6', 'sex', 'canpurpose_dv', 'use_modes_smoke']]
X2 = sm.add_constant(X2)
y2 = reg_data2['work_use_binary']
weights2 = reg_data2['sample_weight']
 
reg_data2['normalized_weight'] = (
    reg_data2['sample_weight'] / reg_data2['sample_weight'].mean()
)

model2 = sm.GLM(
    y2, X2,
    family=sm.families.Binomial(),
    var_weights=reg_data2['normalized_weight']
).fit()
 
odds_ratios2 = pd.DataFrame({
    'OR': np.exp(model2.params),
    'Lower 95% CI': np.exp(model2.conf_int()[0]),
    'Upper 95% CI': np.exp(model2.conf_int()[1]),
    'p-value': model2.pvalues
})
print(odds_ratios2.round(3))
 
print("\nAnalysis complete! These results are now weighted and adjusted.")
```
